```python
train_data = pd.read_csv('../input/network/train_data.csv')
train_data.head()
```

```python
train_data['cont_1'].value_counts()
train_data.shape
```

```python
train_data[train_data.isnull().any(axis=1)]
y = train_data.target
train_data = train_data.drop(['connection_id','target'] , axis=1)
train_data.shape
```

```python
test_data = pd.read_csv('../input/network/test_data.csv')
```

```python
test_ids = test_data.connection_id
test_data.shape
test_data = test_data.drop(['connection_id'], axis=1)
```

```python
from xgboost import XGBClassifier
my_model = XGBClassifier()
my_model.fit(train_data , y)
```

```python

predictions = my_model.predict(test_data)
```

```python
sub = pd.DataFrame({'connection_id': test_ids, 'target': predictions})
sub.head()
```

```python
sub.to_csv('return.csv', index = False)
```
